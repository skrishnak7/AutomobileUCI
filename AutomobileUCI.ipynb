```python
import pandas as pd
import numpy as np
import csv
```

```python
data=[]
with open('/Users/Others/Downloads/imports-85.data') as file:
    rcsv = csv.reader(file, delimiter=',')
    for row in rcsv:
        data.append(row)

df=pd.DataFrame(data, columns = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price'])
df
```

```python
df = df.replace('?', np.nan)
df['num-of-doors']=df.fillna('four')

df['normalized-losses'] = df['normalized-losses'].astype('float64')
df['normalized-losses'].fillna(df['normalized-losses'].mean(),inplace=True)

df['bore'] = df['bore'].astype('float64')
df['bore'].fillna(df['bore'].mean(),inplace=True)

df['stroke'] = df['stroke'].astype('float64')
df['stroke'].fillna(df['stroke'].mean(),inplace=True)

df['peak-rpm'] = df['peak-rpm'].astype('float64')
df['peak-rpm'].fillna(df['peak-rpm'].mean(),inplace=True)

df['horsepower'] = df['horsepower'].astype('float64')
df['horsepower'].fillna(df['horsepower'].mean(),inplace=True)

df['price'] = df['price'].astype('float64')
df['price'].fillna(df['price'].mean(),inplace=True)
```

```python
columns = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
cols_to_transform = ['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']
df1 = pd.get_dummies(df,columns = cols_to_transform)
df1
```

```python
X = df1.drop('price', axis=1)
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
X1=normalize(X)
X2=normalize(df1)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm
```

```python
Y= df1['price']
Y1= X2['price']
```

```python
lm.fit(X2,Y1)
```
